```python
import pandas as pd 
```

```python
clip_gt_pt1_gpt = pd.read_csv("/share/ju/urbanECG/output/street_flooding/train_set_w_clip_gt_pt1_gpt.csv")
```

```python
clip_gt_pt1_gpt.columns
```

```python
# strip the leading and trailing spaces from gpt_classification
clip_gt_pt1_gpt['gpt_classification'] = clip_gt_pt1_gpt['gpt_classification'].str.strip()
```

```python
clip_gt_pt1_gpt['agreed'] = clip_gt_pt1_gpt.gpt_classification == 'flooded road'
```

```python
clip_gt_pt1_gpt['agreed'].value_counts()
```

```python
# randomly sample len(agreed = True) from agreed = False
clip_gt_pt1_gpt_false = clip_gt_pt1_gpt[clip_gt_pt1_gpt['agreed'] == False]
clip_gt_pt1_gpt_true = clip_gt_pt1_gpt[clip_gt_pt1_gpt['agreed'] == True]

clip_gt_pt1_gpt_false_sample = clip_gt_pt1_gpt_false.sample(n=len(clip_gt_pt1_gpt_true), random_state=1)
```

```python
clip_gt_pt1_gpt_false_sample
```

```python
# combine samples 
clip_gt_pt1_gpt_sample = pd.concat([clip_gt_pt1_gpt_true, clip_gt_pt1_gpt_false_sample])
```

```python
# 70, 30 split 
from sklearn.model_selection import train_test_split

train, test = train_test_split(clip_gt_pt1_gpt_sample, test_size=0.3, random_state=1)
```

```python
train
```

```python
# symlink all the files in the train set to /share/ju/urbanECG/static/data/flooding_agreement_model/train/flooded
import os

# symlink agreed images to /share/ju/urbanECG/static/data/flooding_agreement_model/train/flooded 
os.makedirs('/share/ju/urbanECG/static/data/flooding_agreement_model/train/flooded', exist_ok=True)
agreed = train[train['agreed'] == True]

for index, row in agreed.iterrows():
    os.symlink(row['img_path'], '/share/ju/urbanECG/static/data/flooding_agreement_model/train/flooded/' + row['img_path'].split('/')[-1])

# symlink disagreed images to /share/ju/urbanECG/static/data/flooding_agreement_model/train/drivable
os.makedirs('/share/ju/urbanECG/static/data/flooding_agreement_model/train/drivable', exist_ok=True)

disagree = train[train['agreed'] == False]

for index, row in disagree.iterrows():
    os.symlink(row['img_path'], '/share/ju/urbanECG/static/data/flooding_agreement_model/train/drivable/' + row['img_path'].split('/')[-1])
    
```

```python
# symlink all the files in the test set to /share/ju/urbanECG/static/data/flooding_agreement_model/val/flooded
os.makedirs('/share/ju/urbanECG/static/data/flooding_agreement_model/val/flooded', exist_ok=True)

agreed = test[test['agreed'] == True]

for index, row in agreed.iterrows():
    os.symlink(row['img_path'], '/share/ju/urbanECG/static/data/flooding_agreement_model/val/flooded/' + row['img_path'].split('/')[-1])

# symlink disagreed images to /share/ju/urbanECG/static/data/flooding_agreement_model/val/drivable
os.makedirs('/share/ju/urbanECG/static/data/flooding_agreement_model/val/drivable', exist_ok=True)

disagree = test[test['agreed'] == False]

for index, row in disagree.iterrows():
    os.symlink(row['img_path'], '/share/ju/urbanECG/static/data/flooding_agreement_model/val/drivable/' + row['img_path'].split('/')[-1])
```
